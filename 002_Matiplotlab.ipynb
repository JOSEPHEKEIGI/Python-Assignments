### Solution

#### Answer 1 : Loading dataset and explorating data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



data_path = r"jumta.csv"

data = pd.read_csv(data_path)

print(data.head())
```

#### Answer 2:  line plot showing the monthly revenue over the two-year period

```python
y = data['Revenue']
x = data['Month']

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set_title("Monthly Revenue over two years")
ax.set_ylabel("Revenue in KSHS")
ax.set_xlabel("Period in Months")


plt.xticks(rotation=90)
#plt.yticks()

#plt.legend()

plt.show() 
```

#### Answer 3:  Total number of orders per month

```python
y = data['Orders']
x = data['Month']

fig, ax = plt.subplots()
ax.bar(x,y)

ax.set_title(" Total number of orders per month")
ax.set_ylabel("Orders Count")
ax.set_xlabel("Period in Months")


plt.xticks(rotation=90)
#plt.yticks()

#plt.legend()

plt.show() 
```

#### Answer 4: Scatter plot showing the relationship between the number of orders and the average order value

```python
y = data['Average_Order_Value']
x = data['Orders']

fig, ax = plt.subplots()
ax.scatter(x,y)

ax.set_title(" Total number of orders per month")
ax.set_ylabel("Orders Count")
ax.set_xlabel("Period in Months")


plt.xticks(rotation=90)
#plt.yticks()

#plt.legend()

plt.show() 
```
